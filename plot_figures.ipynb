```python
import matplotlib.pyplot as plt
import numpy as np
```

# Acc vs. Prefix Len

## Multi-NLI

```python
species = ("4", "16", "64")
penguin_means = {
    'BERT Prefix': (74.73, 76.87, 77.69),
    'BERT Prefix Gated': (74.76, 76.87, 78.02),
    'ROBERTA Prefix': [79.55, 80.29, 61.97],
    'ROBERTA Prefix Gated': [79.95, 80.10, 82.02]
}

x = np.arange(len(species))  # the label locations
width = 0.12  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    # ax.bar_label(rects, padding=3, fmt='%.1f')  # Format labels to 2 decimal places
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Accuracy (%)')
ax.set_xlabel('Prefix Length')
ax.set_title('MultiNLI')
ax.set_xticks(x + width, species)
# ax.legend(loc='upper left', ncols=3)
ax.legend(loc=(0.02, 0.02))
ax.set_ylim(60, 85)

plt.show()
```

```python
# bert

prefix_lens = [4, 16, 64]
bert_pt_accs = [0.7473, 0.7687, 0.7769]
bert_pt_gated_accs = [0.7476, 0.7687, 0.7802]

plt.plot(prefix_lens, bert_pt_accs, label="bert_pt")
plt.plot(prefix_lens, bert_pt_gated_accs, label="bert_pt_gated")

plt.xlabel("Prefix Length")
plt.ylabel("Accuracy")
plt.title("BERT on Multi-NLI")
plt.legend()
```

```python
# roberta

prefix_lens = [4, 16, 64]
roberta_pt_accs = [0.7955, 0.8029, 0.6197]
roberta_pt_gated_accs = [0.7995, 0.8010, 0.8202]

plt.plot(prefix_lens, roberta_pt_accs, label="roberta_pt")
plt.plot(prefix_lens, roberta_pt_gated_accs, label="roberta_pt_gated")

plt.xlabel("Prefix Length")
plt.ylabel("Accuracy")
plt.title("ROBERTA on Multi-NLI")
plt.legend()
```
